```python
import requests 
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging  
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone%2012pro" # I am able to parametrize "iphone%2012pro" part.

# So whenever we create an app, we create a webpage and through webpage as an input we will pass the last section ("iphone%2012pro") and rest of the part will remain same.
```

```python
flipkart_url 
```

### How to open the above url or any url with python code.

```python
# for that purpose we have to import urlopen from urlib.request

urlclient = urlopen(flipkart_url )
```

### to find out which link we have to click after opening the url use (.read()) method

```python
flipkart_page = urlclient.read()

# it will give the html code of the source page that is the filpkart url we have opened using (urlopen)
```

```python
# but we have to fetch the urls in which we can click one by ones like showed in flipkart after searching iphone12
# so we ahve to  make the dumped data of "urlclient.read()" in a structured format. For that beautifulsoup comes into picture.
```

```python
flipkart_html = bs(flipkart_page, 'html.parser')
```

```python
# Select & copy the "href" (hyperlink from the page inspection and paste it here and add the first part as a string.

"https://www.flipkart.com/" + "apple-iphone-11-black-128-gb/p/itm8244e8d955aba?pid=MOBFWQ6BKRYBP5X8&lid=LSTMOBFWQ6BKRYBP5X8X0KYUG&marketplace=FLIPKART&q=iphone+11&store=tyy%2F4io&srno=s_1_3&otracker=search&otracker1=search&fm=organic&iid=8f708422-1b8d-47a5-a77e-63c8dd4af514.MOBFWQ6BKRYBP5X8.SEARCH&ppt=hp&ppn=homepage&ssid=46dru6q1000000001735034589508&qH=f6cdfdaa9f3c23f3"
```

```python
### Now we have to find out the divisions of the webpage, divisions denotes each different options came on the screen after searching iphone in flipkart

bigbox = flipkart_html.find_all("div" ,{"class"  : "cPHDOP col-12-12"})
```

```python
# Now we want to get how many products are in the bigbox
len(bigbox)
```

```python
# delete first three bigboxes
del bigbox[0:3]
```

```python
# now we need to get the "href url" from the bigboxes so that we can reach to the specific product.
# So now we focus only on first bigbox, (bigbox[0]), then div of div of div of a

bigbox[0].div.div.div.a["href"]
```

```python
# then append the flipkart url with this to get the whole url.

productlink = "https://www.flipkart.com" + bigbox[0].div.div.div.a["href"]
productlink # now this leads us to inside of a single product.
```

```python
# Now this is for a sigle bigbox, but we have 29 bigboxes so we will use the automated for loop

for i in bigbox:
    print( "https://www.flipkart.com" + i.div.div.div.a["href"])

# Now we got link for each product.
```

```python
# now we have to get into a particular product and extract all the details inside that. 
product_req = requests.get(productlink)
```

```python
# Now how to read that 
product_req.text
```

```python
# again it is a dump, to get it in human readable format we have to use beautysoup format

product_html = bs(product_req.text, 'html.parser')
product_html
```

```python
# Now out of this entire html we have to scrap only the reviews, ratings etc.
# For that again we have to go to div then specific class because all the reviews have same class

comment_box = product_html.find_all("div", {"class":"RcXBOT d7A196"})

# so now it is giving all the bigboxes of the reviews of a specific product bigbox[0]
```

```python
# Now we check how many comments we have.

len(comment_box)
```

```python
# Now the next part is we want to know the persons name who has given the commands.

comment_box[0].div.div.find_all("p",{"class":"_2NsDsF AwS1CA"})[0].text
```

```python
# Now to get all the names we have to use loop

for i in comment_box:
    print(i.div.div.find_all("p",{"class":"_2NsDsF AwS1CA"})[0].text, end ="\n")
    
```

```python
# Now we have to find the ratings out of 5

for i in comment_box:
    print(i.div.div.div.div.text)
```

```python
# Now we want to extract the header like "awesome", "perfect product" etc.

for i in comment_box:
    print(i.div.div.div.p.text)
```

```python
# Now we have to extract the comments
for i in comment_box:
    print(i.div.div.find_all("div",{"class":""})[0].text)
```
